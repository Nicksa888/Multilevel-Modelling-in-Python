# Simple (Intercept Only) Multilevel Models 

Here, we want to determine the extent to which vocabulary scores can be used to predict general reading achievement. Students are nested within schools, which means that standard linear regression models are not appropriate. Here, school is a random effect and vocabulary scores are fixed. The first model to be fitted is the null model with no independent variable. This model is useful for obtaining estimates of the residual and intercept variance when only the clustering by school is considered

```python
model = smf.mixedlm("geread ~ 1", Achieve, groups = Achieve['school'])
result = model.fit()
# Print the summary of the model
print(result.summary())
```

```python
# Store the model results in a list
model_results = [result]

# Extract the relevant information from the model results
model_summaries = []
for idx, model_result in enumerate(model_results):
    model_info = {
        'Model': f"Model {idx+1}",
        'Fixed_Effect': model_result.params['Intercept'],
        'Fixed_Effect_SE': model_result.bse['Intercept'],
        'Random_Effect': model_result.cov_re.iloc[0, 0] ** 0.5,
    }
    model_summaries.append(model_info)

# Create a DataFrame from the summaries
summary_df = pd.DataFrame(model_summaries)

# Print the summary table
print(summary_df)
```

```python
# Fit the random intercept model with 'school' as a random effect and 'gevocab' as a fixed effect
model = smf.mixedlm("geread ~ gevocab", Achieve, groups = Achieve['school'])
result = model.fit()
# Print the summary of the model
print(result.summary())
```

The estimated coefficient for the 'gevocab' predictor is 0.513. This indicates that, on average, for each one-unit increase in 'gevocab', the general reading achievement score is expected to increase by 0.513 units.

# Create Three Level Models

```python
# Define the three level m
import pandas as pd
import statsmodels.api as sm

# Load the data into a pandas DataFrame (assuming the data is stored in the 'Achieve' DataFrame)
Achieve = pd.DataFrame(Achieve)

# Convert 'school' and 'class' columns to strings and create a new unique identifier for the nested structure
Achieve['school_class'] = Achieve['school'].astype(str) + '_' + Achieve['class'].astype(str)

# Define the three-level null model
model_formula = "geread ~ 1"
group_var = 'school_class'

# Fit the three-level null model
model = sm.MixedLM.from_formula(model_formula, data=Achieve, groups=Achieve[group_var])
result = model.fit()

# Print the summary of the model
print(result.summary())
```

```python
# Rename the 'class' column to 'class_id' in the DataFrame
Achieve.rename(columns={'class': 'class_id'}, inplace=True)

# Model formula: geread ~ school + class + other independent variables
# Make sure to include all relevant independent variables in the formula
S_C_model_formula = 'geread ~ school + class_id'
# Fit the three-level model
S_C_model = smf.mixedlm(S_C_model_formula, Achieve, groups = Achieve['class_id'], re_formula = "~school")
S_C_result = S_C_model.fit(maxiter = 2000, # Increase the number of iterations
                           method = 'nm') # Try a different optimization method (e.g., 'nm' for Nelder-Mead)
```

```python
# View the model summary
print(S_C_result.summary())
```

S_C_Model Interpretation:

Fixed Effects:

Intercept: The intercept represents the average reading score (geread) when all the predictor variables (school and class_id) are zero. In this model, the estimated intercept is 4.468. The p-value (P>|z|) is very small (close to 0), indicating that the intercept is significantly different from zero. This value represents the average reading score when both school and class_id are 0.

school: The variable school represents the effect of the school on reading scores (geread). The coefficient estimate is very close to 0 (0.000), and the p-value is 0.999. This means that the effect of school on reading scores is not statistically significant at the conventional significance level (usually 0.05). In other words, the school variable does not have a meaningful impact on the reading scores.

class_id: The variable class_id represents the effect of the class on reading scores (geread). The coefficient estimate is -0.011, and the p-value is 0.833. Like school, the class_id variable is also not statistically significant, suggesting that class membership does not have a significant impact on reading scores.

Random Effects:

Group Var: The Group Var represents the variance of the random effect at the group level. In this model, the estimated variance is 3.842. The random effect captures the variability in reading scores between different groups (likely representing schools or school classes).

Group x school Cov: This represents the covariance between the random effect and the school variable. The value is 0.016. The covariance indicates the relationship between the school-level random effect and the school variable.

school Var: The school Var represents the variance of the random effect at the school level. The estimated variance is 0.000. A variance of 0 for the school-level random effect might indicate that there is very little variability between different schools, suggesting that schools are very similar to each other with respect to reading scores.

Overall, the model suggests that neither the school variable nor the class_id variable has a significant impact on the reading scores (geread). The significant intercept indicates that there is a baseline reading score when both school and class_id are zero, which might represent the overall average reading score in the data. However, the random effects (Group Var, Group x school Cov, school Var) show that there is some variability in reading scores between different groups (schools or classes).

Next, we want to model the raandom effects are specified using (1 | school/class). This means that we have a random intercept for each combination of school and class.


```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Convert 'school' and 'class_id' columns to strings
Achieve['school'] = Achieve['school'].astype(str)
Achieve['class_id'] = Achieve['class_id'].astype(str)

# Model formula: geread ~ school + class_id + other independent variables
# Make sure to include all relevant independent variables in the formula
S_C_model_formula = 'geread ~ school + class_id'

# Fit the three-level model with random intercepts for each school/class combination
S_C_model = smf.mixedlm(S_C_model_formula, data=Achieve, groups=Achieve['school'] + '/' + Achieve['class_id'])
S_C_result = S_C_model.fit()

# Print the model summary
print(S_C_result.summary())

```

The format of the random effects is a dictionary where the keys are the school/class combinations (e.g., '1033/1', '1075/1', '1103/1', etc.), and the values are the estimated random intercepts for each corresponding school/class combination.
